# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Calculate the totals, averages, and percentages in the DataFrame
total_schools = len(school_data_complete["school_name"].unique())
total_students = school_data_complete["student_name"].count()
total_budget = school_data["budget"].sum()
avg_math_score = round(school_data_complete["math_score"].mean(),2)
avg_reading_score = round(school_data_complete["reading_score"].mean(), 2)
only_passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70, :]
only_passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70, :]
percent_math = round((len(only_passing_math) / total_students) * 100, 2)
percent_reading = round((len(only_passing_reading) / total_students) * 100, 2)
overall_passing = round((avg_math_score + avg_reading_score) / 2, 2)
```

```python
# Group the complete dataset by school
groupby_schools = school_data_complete.groupby(["school_name"])
```

```python
# Calculate all the values for each column
top_school_type = school_data.set_index(["school_name"])["type"]
top_total_students = groupby_schools["student_name"].count()
top_total_budget = groupby_schools["budget"].sum()
top_total_students = groupby_schools["student_name"].count()
top_total_budget = groupby_schools["budget"].mean()
per_student_budget = groupby_schools["budget"].mean() / groupby_schools["size"].count()
top_avg_math_score = round(groupby_schools["math_score"].mean(), 2)
top_avg_reading_score = round(groupby_schools["reading_score"].mean(), 2)
top_only_passing_math_count = only_passing_math["school_name"].value_counts()
top_only_passing_reading_count = only_passing_reading["school_name"].value_counts()
top_percent_math = round((top_only_passing_math_count / top_total_students) * 100, 2)
top_percent_reading = round((top_only_passing_reading_count / top_total_students) * 100, 2)
top_overall_passing = round((top_percent_math + top_percent_reading) / 2, 2)
```

```python
top_school_type
```

```python
# School Summary table created into DataFrame
school_summary_table = pd.DataFrame({"School Type": top_school_type,
                                  "Total Students": top_total_students,
                                  "Total School Budget": top_total_budget,
                                  "Per Student Budget": per_student_budget,
                                  "Average Math Score": top_avg_math_score,
                                  "Average Reading Score": top_avg_reading_score,
                                  "% Passing Math": top_percent_math,
                                  "% Passing Reading": top_percent_reading,
                                  "% Overall Passing Rate": top_overall_passing})
school_summary_table
```

```python
# # Merge all the summary data together
school_data_merged = pd.merge(school_data_complete, school_summary_table, how="left", on=["% Overall Passing Rate", "% Overall Passing Rate"])
school_data_merged.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
math_score_9th = school_data_complete.loc[school_data_complete["grade"] == "9th"].groupby("school_name", as_index=False)
math_score_10th = school_data_complete.loc[school_data_complete["grade"] == "10th"].groupby("school_name", as_index=False)
math_score_11th = school_data_complete.loc[school_data_complete["grade"] == "11th"].groupby("school_name", as_index=False)
math_score_12th = school_data_complete.loc[school_data_complete["grade"] == "12th"].groupby("school_name", as_index=False)
```

```python
math_avg_9th = round(pd.DataFrame(math_score_9th["math_score"].mean()), 2)
math_avg_10th = round(pd.DataFrame(math_score_10th["math_score"].mean()), 2)
math_avg_11th = round(pd.DataFrame(math_score_11th["math_score"].mean()), 2)
math_avg_12th = round(pd.DataFrame(math_score_12th["math_score"].mean()), 2)
```

```python
# Merge all the data together
math_avg_by_grade = pd.merge(math_avg_9th, math_avg_10th, on="school_name")
math_avg_by_grade = pd.merge(math_avg_by_grade, math_avg_11th, on="school_name")
math_avg_by_grade = pd.merge(math_avg_by_grade, math_avg_12th, on="school_name")
math_avg_by_grade.columns = ["School Name","9th Grade","10th Grade","11th Grade","12th Grade"]
math_avg_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
binlabels = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# top_summary_table = top_summary_table.reset_index(drop=True)
school_summary_table["Spending Ranges"] = pd.cut(school_summary_table["Per Student Budget"], spending_bins, labels=binlabels)
school_summary_table
```

```python
# Creating a group based off of the bins             
budget_ranges = school_summary_table[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate", "Spending Ranges"]]
budget_ranges = round(budget_ranges.groupby(["Spending Ranges"]).mean(), 2)
budget_ranges
```
